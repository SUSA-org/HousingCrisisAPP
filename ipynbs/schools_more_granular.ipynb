```python
import pandas as pd
import geopandas as gpd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
crimes = pd.read_csv("csvs/Crimes_and_Clearances_with_Arson-1985-2016.csv")
schools = pd.read_csv("csvs/new_district.csv")
places = pd.read_csv("csvs/htaindex_data_places_06.csv")
places_info = gpd.read_file("tl_2017_06_place/tl_2017_06_place.shp")
```

```python
crimes.columns
```

```python
crimes = crimes[crimes['Year'] == 2016]
```

```python
crimes.head()
```

```python
len(crimes)
```

```python
crimes_by_city = crimes.groupby("NCICCode").aggregate(sum).reset_index()
crimes_by_city.head()
```

```python
len(crimes_by_city)
```

```python
crimes_by_city = crimes_by_city[['NCICCode', 'Violent_sum', 'Property_sum']]
crimes_by_city
```

```python
places['place'] = places['place'].apply(lambda k: k[1:-1])
```

```python
places['name'] = places['name'].apply(lambda k: k[1:-1])
```

```python
print(places.columns)
places.head()
```

```python
print(places_info.columns)
places_info.head()
```

```python
#check validity
places_info[places_info['NAME'] == 'Dublin']
```

```python
places_geo = places_info[['GEOID', 'geometry']]
places_geo.head()
```

```python
hta_with_geo = pd.merge(places, places_geo, how='left', left_on='place', right_on='GEOID')
hta_with_geo.drop('GEOID', axis=1, inplace=True)
hta_with_geo.head()
```

```python
schools_with_crimes = pd.merge(schools, crimes_by_city, how='left', left_on='City', right_on='NCICCode')
schools_with_crimes.drop(['NCICCode', 'Unnamed: 0'], axis = 1, inplace=True)
schools_with_crimes.head()
```

```python
schools_full = pd.merge(schools_with_crimes, hta_with_geo, how = 'left', left_on='City', right_on='name')
schools_full.drop('name', axis=1, inplace=True)
schools_full.head()
```
