```python
counties = gpd.read_file("../tl_2017_us_county/tl_2017_us_county.shp")
counties.head()
```

```python
ca_counties = counties[counties['STATEFP'] == '06']
ca_counties.drop("geometry", axis=1, inplace=True)
ca_counties.head()
```

```python
add_county_to_ht_bg = pd.merge(test, ca_counties[["NAME", "COUNTYFP"]], how='left', on='COUNTYFP')
```

```python
add_county_to_ht_bg.head()
```

```python
add_county_to_ht_bg.drop(["GEOID", "COUNTYFP"], axis = 1, inplace=True)
add_county_to_ht_bg.rename(columns={"blkgrp":"GEOID"}, inplace=True)
add_county_to_ht_bg.head()
```

```python
add_county_to_ht_bg.columns
```

```python
blk.columns
```

```python
crimes.drop("Unnamed: 0", axis=1, inplace=True)
crimes.head()
```

```python
schools_by_county = schools[['County', 'Rank score (2017)']].groupby("County").aggregate(np.nanmean).reset_index()
schools_by_county.head()
```

```python
crimeschools = pd.merge(crimes, schools_by_county, on="County")
crimeschools.head()
```

```python
bg_full = pd.merge(add_county_to_ht_bg, crimeschools, how = 'left', left_on='NAME', right_on='County')
bg_full.drop("NAME", axis=1, inplace=True)
bg_full.head()
```
