```python
import pandas as pd
```

```python
es = pd.read_csv("CAElementarySchool.csv")
ms = pd.read_csv("CAMiddleSchool.csv")
hs = pd.read_csv("CAHighSchool.csv")
district = pd.read_csv("CADistrictSchools.csv")
city = pd.read_csv("CACitiesSchools.csv")
```

```python
#district.rename(columns={"Rank (of 849)":"Rank"}, inplace=True)
```

```python
#district["rank"] = district["Rank (of 849)"].fillna("0th").apply(lambda k: k[:len(k)-2])
district["rank"] = district["Rank"].fillna("0th").apply(lambda k: k[:len(k)-2])
```

```python
# district.drop(["test"], axis=1, inplace=True)
district.head()
```

```python
district[district["District"].apply(lambda k: "Du" in k)].sort_values("District")
```

```python
hs_avg = hs.groupby("District").mean().sort_values("Average Standard Score (2017)", ascending = False).reset_index()
hs_avg = hs_avg.rename(columns={"Average Standard Score (2017)":"Avg HS Score"})
hs_avg.head()
```

```python
ms_avg = ms.groupby("District").mean().sort_values("Average Standard Score (2017)", ascending = False).reset_index()
ms_avg = ms_avg.rename(columns={"Average Standard Score (2017)":"Avg MS Score"})
ms_avg.head()
```

```python
es_avg = es.groupby("District").mean().sort_values("Average Standard Score (2017)", ascending = False).reset_index()
es_avg = es_avg.rename(columns={"Average Standard Score (2017)":"Avg ES Score"})
es_avg.head()
```

```python
district_avg = district.groupby("District").mean().sort_values("Rank score (2017)", ascending = False).reset_index()
district_avg.head()
```

```python
city.head()
```

```python
print(sorted(set(city.City)))
```

```python
total = pd.merge(hs_avg, ms_avg, how='inner', on="District")
total = total[["District", "Avg HS Score", "Avg MS Score"]]
total.head()
```

```python
total = pd.merge(total, es_avg, how='inner', on="District")
total = total[["District", "Avg HS Score", "Avg MS Score", "Avg ES Score"]]
total.head()
```

```python
total["avg"] = (total["Avg HS Score"] + total["Avg MS Score"] + total["Avg ES Score"]) / 3
```

```python
temp = pd.merge(total, district, how = 'inner', on = 'District')
temp = temp[["District", "Avg HS Score", "Avg MS Score", "Avg ES Score", "avg", "Rank score (2017)"]]
temp.sort_values("avg", ascending=False, inplace=True)
temp["Rank score (2017)"] = temp['Rank score (2017)'] * 100
temp.to_csv("compare.csv")
temp.head()
```
